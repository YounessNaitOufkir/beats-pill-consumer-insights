# **PART I: Data Gathering**

```python
# Importing the necessary libraries
import pandas as pd
import requests
from glob import glob
```

```python
# Structure payload.
payload = {
   'source': 'amazon_reviews',
   'query': 'B0D4SX9RC6',
   'parse': True,
   'domain': 'com',
   'start_page': '9',
   'pages': '20',
   'language': 'en'
}

# Get response.
response = requests.request(
    'POST',
    'https://realtime.oxylabs.io/v1/queries',
    auth=('Oxylabs API'),
    json=payload,
)

results = response.json()['results']
```

```python
reviews_data = []
count = 0
for result in results:
  asin = result['content']["asin"]
  for review in result['content']['reviews']:
    count += 1
    try:
      review_info = {
      'review_id': review['id'],
      'product_id': asin,
      "title": review["title"],
      "rating": review["rating"],
      "content": review["content"],
      "timestamp": review["timestamp"],
      "is_verified": review["is_verified"],
      "helpful_count": review.get('helpful_count', 0)
      }
      reviews_data.append(review_info)
    except:
      print(count)
      break

reviews_df = pd.DataFrame(reviews_data)
reviews_df
```

```python
reviews_df.to_csv('reviews_beats.csv', index=False)
```

```python
# creating a list that contains the filenames for all the reviews csv files
files = glob('reviews_*.csv')
files
```

```python
dfs = []

# Reading each CSV file into a DataFrame and append it to the empty list
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)
#merged_df
```

```python
# checking for duplicates
merged_df.duplicated().sum()
```

```python
# remove duplicated
merged_df.drop_duplicates(inplace=True)
```

```python
# exporting the merged data into a csv for the next step
merged_df.to_csv('merged_df_us.csv', index=False)
```
